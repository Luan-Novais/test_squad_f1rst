```
#Realizando a leitura do arquivo 

df = spark.read.option("header","True").option("delimiter",";").csv("dbfs:/FileStore/dados_sivep_esus_sao_paulo_sp_jul_dez_2022.csv")
```

```
#validando a quantidade de registro = 999.054
df.count()
```

```
#Recuperando as colunas
# [ SEXO,DT_SIN_PRI,CS_RACA,FEBRE,TOSSE,HOSPITAL,DT_INTERNA,DT_EVOLUCA ]

from pyspark.sql.functions import *
from pyspark.sql.types import *
df_new = df.select(col("SEXO"),col("DT_SIN_PRI"),col("CS_RACA"),col("FEBRE"),col("TOSSE"),col("HOSPITAL"),col("DT_INTERNA"),col("DT_EVOLUCA"))

df_new.printSchema
```

```
#renomear colunas
# DT_SIN_PRI = DATA_INICO_SINT
# DT_INTERNA = DATA_INTERNACAO
# DT_EVOLUCA = DATA_EVOLUCAO
df_new = df_new.withColumnRenamed("DT_SIN_PRI","DATA_INICO_SINT")\
.withColumnRenamed("DT_INTERNA","DATA_INTERNACAO")\
.withColumnRenamed("DT_EVOLUCA","DATA_EVOLUCAO")\

display(df_new)
```

```
#Realizar regra da coluna teve_febre e realizar uma contagem com base no sexo e se teve febre 

df_new = df_new.withColumn("teve_febre", when(col("FEBRE") == "1", lit("sim"))
               .when(col("FEBRE") == "2", lit("nao")).otherwise("ignorado"))

df_new.select("SEXO","teve_febre").groupBy("sexo","teve_febre").count().orderBy("SEXO").show()

# +----+----------+------+
# |sexo|teve_febre| count|
# +----+----------+------+
# |   F|       nao|  2450|
# |   F|  ignorado|438097|
# |   F|       sim|179034|
# |   M|  ignorado|248124|
# |   M|       nao|  2164|
# |   M|       sim|129185|
# +----+----------+------+
```

```
#Realizar regra da coluna teve_tosse e realizar uma contagem com base no sexo e se teve tosse 

df_new = df_new.withColumn("teve_tosse", when(col("TOSSE") == "1", lit("sim")).otherwise("nao"))
df_new.select("SEXO","teve_tosse").groupBy("sexo","teve_tosse").count().orderBy("SEXO").show()

# +----+----------+------+
# |sexo|teve_tosse| count|
# +----+----------+------+
# |   F|       nao|224614|
# |   F|       sim|394967|
# |   M|       nao|140384|
# |   M|       sim|239089|
# +----+----------+------+
```

```
#Realizar regra da coluna paciente_internado e realizar uma contagem com base no sexo e se paciente_internado 

df_new = df_new.withColumn("paciente_internado", when(col("HOSPITAL") == "1", lit("sim")).otherwise("nao"))
df_new.select("SEXO","paciente_internado").groupBy("sexo","paciente_internado").count().orderBy("SEXO").show()


# +----+------------------+------+
# |sexo|paciente_internado| count|
# +----+------------------+------+
# |   F|               nao|611428|
# |   F|               sim|  8153|
# |   M|               nao|371475|
# |   M|               sim|  7998|
# +----+------------------+------+
```

```
#Mostrar somente paciente que foram internado com SEXO ,data de DATA_INTERNACAO e DATA_EVOLUCAO recuperar outro df com ods campos e tranformar os campos de data para o tipo date 

#Qt_pacientes_internado = 16.151
df_pac_internados = df_new.filter(col("paciente_internado")=="sim").select(col("SEXO"),col("DATA_INTERNACAO"),col("DATA_EVOLUCAO"))
```

```
#Calcular os dias em que a doença teve evolução 
df_pac_internados = df_pac_internados.withColumn("dias_evolucao", datediff(col("DATA_INTERNACAO"), col("DATA_EVOLUCAO"))*-1)

# +----+---------------+-------------+-------------+
# |SEXO|DATA_INTERNACAO|DATA_EVOLUCAO|dias_evolucao|
# +----+---------------+-------------+-------------+
# |   M|     2022-06-30|   2022-07-03|            3|
# |   M|     2022-06-29|   2022-06-30|            1|
# |   M|     2022-06-15|   2022-06-23|            8|
# |   M|     2022-06-14|   2022-06-22|            8|
# |   F|     2022-06-27|   2022-07-15|           18|
# +----+---------------+-------------+-------------+



```

```
#Mostrar dias_evolucao maior por sexo

# +----+-------------------+
# |SEXO|dias_evolucao_maior|
# +----+-------------------+
# |   F|                696|
# |   M|                852|
# +----+-------------------+
df_maior = df_pac_internados.select(col("SEXO"),col("dias_evolucao")).groupBy("SEXO").agg(max(col("dias_evolucao")).alias("dias_evolucao_maior"))
df_maior.show()
```

```
#Mostrar dias_evolucao menor por sexo > 0
# +----+-------------------+
# |SEXO|dias_evolucao_menor|
# +----+-------------------+
# |   F|                  1|
# |   M|                  1|
# +----+-------------------+


df_menor = df_pac_internados.filter(col("dias_evolucao") > 0).select(col("SEXO"),col("dias_evolucao")).groupBy("SEXO").agg(min(col("dias_evolucao")).alias("dias_evolucao_menor"))
df_menor.show()
```

```
#criar regra de classificacao de raca e contar a qt de pessoas por raca e por sexo remover null
df_new = df_new.withColumn("RACA", when(col("CS_RACA") == "1", lit("branca"))\
                           .when(col("CS_RACA") == "2", lit("preta"))\
                           .when(col("CS_RACA") == "3", lit("amarela"))\
                           .when(col("CS_RACA") == "4", lit("parda"))\
                           .when(col("CS_RACA") == "5", lit("indigena"))\
                           .when(col("CS_RACA") == "9", lit("ignorado")))

# +--------+----+------+
# |    RACA|SEXO| count|
# +--------+----+------+
# |indigena|   M|   161|
# |   preta|   M| 18294|
# |ignorado|   F|179121|
# | amarela|   F| 14055|
# |  branca|   M|117948|
# |indigena|   F|   216|
# |ignorado|   M|111490|
# |   preta|   F| 28095|
# | amarela|   M|  8799|
# |  branca|   F|202608|
# |   parda|   M|122777|
# |   parda|   F|195482|
# +--------+----+------+

df_new.filter(df_new.RACA.isNotNull()).select(col("RACA"),col("SEXO")).groupBy(col("RACA"),col("SEXO")).count().show()
```
